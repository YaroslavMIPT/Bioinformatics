Анализ отчета QUAST показывает, что сборка SPAdes значительно превосходит сборки Velvet. Геном собрался в более длинные и цельные фрагменты. Самый длинный контиг (Largest contig) у SPAdes также больше.

Velvet оказался очень чувствителен к размеру k-mer. SPAdes выигрывает за счет того, что использует сразу несколько k-mer и имеет встроенный модуль коррекции ошибок (mismatch corrector).

## Часть 3: Улучшение сборки
Соображения по улучшению параметров:

Для Velvet: Добавление параметров -exp_cov auto и -cov_cutoff auto позволит алгоритму автоматически вычислить ожидаемое покрытие и отсечь контиги с аномально низким покрытием. Это должно очистить сборку от лишних контигов, возникших из-за ошибок секвенирования.

Для SPAdes: Добавление более длинного k-mer (99) позволит алгоритму лучше разрешать длинные повторяющиеся участки генома, что потенциально может сшить короткие контиги в более длинные скаффолды.

Сравнение в QUAST подтвердило ожидания. Улучшенная сборка Velvet стала значительно чище: количество контигов резко сократилось, так как отфильтровался мусор, а метрика N50 возросла. Улучшенный SPAdes также лежит в репозитории.

## Итоговые выводы
Для сборки вирусных геномов de novo SPAdes является более мощным и автоматизированным инструментом, так как его мульти-k-mer-ный подход максимизирует длину контигов. Velvet является хорошим классическим ассемблером, но для получения качественного результата он требует тонкой ручной настройки параметров покрытия и длины k-mer.
